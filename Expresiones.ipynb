## Precedencia
La precedencia (prioridad) indica el orden en que se eval√∫an los operadores.

La precedencia de los operadores es similar a la algebraica:

* Primero, negaci√≥n aritm√©tica (`-`)
* Enseguida, exponenciaci√≥n (`**`)
* Despu√©s, multiplicaci√≥n y divisi√≥n (`*`, `/`, `//`, `%`)
* Al √∫ltimo, sumas y restas (`+`, `-`)

Puede alterse el orden usando par√©ntesis. Las operaciones entre par√©ntesis siempre tienen precedencia.

```python
2 + 7 * 3 - 8
```

```python
(2 + 7) * (3 - 8)
```

## Asociatividad

La asociatividad de los operadores se refiere al orden de evaluaci√≥n cuando en la misma expresi√≥n concurren operadores de la misma precedencia.

- La asociatividad es de izquierda a derecha para todos los operadores, excepto el de exponenciaci√≥n
- La asociatividad del operador de exponenciaci√≥n es de derecha a izquierda.

```python
2 ** 3 ** 2
```

```python
(2 ** 3) ** 2
```

```python
2 ** (3 ** 2)
```
